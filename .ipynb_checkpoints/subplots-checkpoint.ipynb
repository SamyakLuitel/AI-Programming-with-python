# subplots

```python
%matplotlib notebook

import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.subplot?
```

```python
plt.figure()
plt.subplot(1,2,1)

linear_data = np.array([1,2,3,4,5,6,7,8])

plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2

plt.subplot(1,2,2)
plt.plot(exponential_data,"-o")
```

```python
# puting exponetial in linear graph

plt.figure()
plt.subplot(1,2,1)

linear_data = np.array([1,2,3,4,5,6,7,8])

plt.plot(linear_data, '-o')


exponential_data = linear_data**2

plt.subplot(1,2,2)
plt.plot(exponential_data,"-o")


plt.subplot(1,2,1)
plt.plot(exponential_data,'-x')
```

#### we see that the axis of the 2 subplots are not locked  so to have subplots sof same scale we have to share axis 

```python
plt.figure()
ax1 = plt.subplot(1,2,1)
plt.plot(linear_data,'-o')
ax2 = plt.subplot(1,2,2, sharey = ax1)
plt.plot(exponential_data,'-x')
```
